```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel.groupby('reservation_status_date')[['adr']].mean()

city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python

resort_hotel_mean = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel_mean = city_hotel.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize=(12, 6))

# Plot for Resort Hotel ADR (mean) over time
plt.plot(resort_hotel_mean.index, resort_hotel_mean['adr'], 
         label='Resort Hotel', color='blue', linewidth=.5)

# Plot for City Hotel ADR (mean) over time
plt.plot(city_hotel_mean.index, city_hotel_mean['adr'], 
         label='City Hotel', color='orange', linewidth=.5)

# Customize the plot with title and labels
plt.title('Average ADR (Mean) Over Time by Hotel Type')
plt.xlabel('Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()

# Show the plot
plt.show()
```

```python
import calendar
df['month']=df['reservation_status_date'].dt.month
df['month'] = df['month'].apply(lambda x: calendar.month_abbr[x])

plt.figure(figsize=(10, 6))

# Plotting the count of reservations and cancellations per month with month names
sns.countplot(x='month', hue='is_canceled', data=df, palette='Blues', order=calendar.month_abbr[1:])

# Customize the plot
plt.title('Monthly Trend of Reservations and Cancellations')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.legend(title='Booking Status', labels=['Not Canceled', 'Canceled'])

# Show the plot
plt.show()
```

```python

# Ensure that the 'month' column is already created
df['month'] = df['reservation_status_date'].dt.month

# Filter data for canceled bookings and group by month, then sum ADR
monthly_adr_sum_canceled = df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()

# Map numeric months to the first three letters of month names
monthly_adr_sum_canceled['month_name'] = monthly_adr_sum_canceled['month'].apply(lambda x: calendar.month_name[x][:3])

# Create a colorful palette for each month
colors = sns.color_palette("Set3", len(monthly_adr_sum_canceled))

# Plotting the sum of ADR for canceled bookings per month with different colors
plt.figure(figsize=(10, 6))

# Create a barplot with unique colors for each month
sns.barplot(x='month_name', y='adr', data=monthly_adr_sum_canceled, palette=colors)

# Customize the plot
plt.title('Total ADR for Canceled Bookings by Month')
plt.xlabel('Month')
plt.ylabel('Total ADR')

# Show the plot
plt.show()
```

```python
# Filter data for canceled bookings
cancelled_data = df[df['is_canceled'] == 1]

# Get top 10 countries with the most canceled reservations
top10_countries = cancelled_data['country'].value_counts()[:10]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top10_countries, labels=top10_countries.index, autopct='%.2f', colors=sns.color_palette("Set3", 10))

# Adjust label rotation for better readability
plt.gca().set_aspect('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Top 10 Countries with Reservation Cancellations')

# Rotate the labels for better visibility
plt.tight_layout()  # Ensures everything fits without clipping.

# Show the plot
plt.show()
```

```python
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date' )
notcancelled_data = df[df['is_canceled'] == 0]

notcancelled_df_adr=notcancelled_data.groupby('reservation_status_date')[['adr']].mean()
notcancelled_df_adr.reset_index(inplace=True)
notcancelled_df_adr.sort_values('reservation_status_date')
plt.figure(figsize=(20, 6))

# Plot the ADR for canceled bookings
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled', color='orange')

# Plot the ADR for non-canceled bookings
plt.plot(notcancelled_df_adr['reservation_status_date'], notcancelled_df_adr['adr'], label='Not Canceled', color='blue')

# Customize the plot
plt.title('Average ADR for Canceled and Non-Canceled Bookings Over Time')
plt.xlabel('Reservation Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()

# Display the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Calculate average ADR for canceled bookings
cancelled_df_adr = df[df['is_canceled'] == 1].groupby('reservation_status_date')[['adr']].mean().reset_index()

# Calculate average ADR for non-canceled bookings
notcancelled_df_adr = df[df['is_canceled'] == 0].groupby('reservation_status_date')[['adr']].mean().reset_index()

# Filter data for the period from 2016 to September 2017 for both canceled and non-canceled ADR
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] >= '2016-01-01') & (cancelled_df_adr['reservation_status_date'] <= '2017-09-30')]
notcancelled_df_adr = notcancelled_df_adr[(notcancelled_df_adr['reservation_status_date'] >= '2016-01-01') & (notcancelled_df_adr['reservation_status_date'] <= '2017-09-30')]

# Sort values by reservation status date for both datasets
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)
notcancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plotting the results
plt.figure(figsize=(20, 6))

# Plot the ADR for canceled bookings in 2016-2017
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Canceled', color='orange', linewidth=1)

# Plot the ADR for non-canceled bookings in 2016-2017
plt.plot(notcancelled_df_adr['reservation_status_date'], notcancelled_df_adr['adr'], label='Not Canceled', color='blue', linewidth=1)
# Customize the plot
plt.title('Average ADR for Canceled and Non-Canceled Bookings (2016 - September 2017)')
plt.xlabel('Reservation Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()

# Display the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
